```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error 
#data reading
df=pd.read_csv("diabetes.csv")  
df.head()
```

```python
x1=df
y1=x1.pop("Outcome")
```

```python
x1_tr,x1_te,y1_train,y1_test=train_test_split(x1,y1,test_size=0.3,random_state=0)

train_status=x1.describe()
train_status=train_status.transpose()
def norm(x):
  return(x-train_status['mean'])/train_status['std'];
x1_train=norm(x1_tr)
x1_test=norm(x1_te)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
%tensorflow_version 1.x
```

```python
def build_model():
  model=keras.Sequential([
                          layers.Dense(64,activation=tf.nn.relu,input_shape=[len(x1_train.keys())]),
                          layers.Dense(64,activation=tf.nn.relu),
                          layers.Dense(1,activation=tf.nn.sigmoid)
  ])
  optimizer=tf.keras.optimizers.RMSprop(0.001)
  model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['binary_crossentropy'])
  return model
model=build_model()
model.summary()
```

```python
class Dots(keras.callbacks.Callback):
  def on_epoch_end(self,epoch,logs):
    if epoch%100==0: print('')
    print('.',end='')

EPOCHS=150
history=model.fit(x1_train,y1_train,epochs=EPOCHS,validation_split=0.2,verbose=0,callbacks=[Dots()])

hist=pd.DataFrame(history.history)
hist['epoch']=history.epoch
hist.tail()
```

```python
pred=model.predict(x1_test)
pred=[1 if y>=0.5 else 0 for y in pred]
```

```python
yp=[1 if y==1 else 0 for y in y1_test]
yp
```

```python
total=0
correct=0
wrong=0
for i in pred:
  total=total+1
  if(yp[i]==pred[i]):
    correct=correct+1
  else:
     wrong=wrong+1
print("total"+str(total))
print("Correct"+str(correct))
print("Wrong"+str(wrong))
```
